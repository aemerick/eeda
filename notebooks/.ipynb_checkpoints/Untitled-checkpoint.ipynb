```python
%matplotlib inline

import numpy as np
import pandas as pd

from learntools.core import binder
```

This notebook will have some rudimentary data exploration from kaggle.com/dansbecker 's intro.... and a way for me to start using pandas

```python
filepath = "./../input/melb_data.csv"
raw_melb_data = pd.read_csv(filepath)
raw_melb_data.describe()
```

```python
melb_data = raw_melb_data.dropna(axis=0) # filter empty 
```

```python
#
# select some features to build the model
#
melbourne_features = ['Rooms','Bathroom','Landsize','Lattitude',
                      'Longtitude'] # why is this spelled wrong in the dataset?
                      

#
# do the selection
#
X = melb_data[melbourne_features]

#
# y variable we would like to predict
#
y = melb_data.Price

print(X.describe())
print(X.head())
```

# Now build some model of this using scikitlearn

```python
from sklearn.tree import DecisionTreeRegressor

# Define the model (random state hard-coded since this is non-deterministic)
melbourne_model = DecisionTreeRegressor(random_state=1)

# Fit
melbourne_model.fit(X,y)
```

```python
print("Testing the model using the training data")
print(X.head())
print("The predictions are :")
print(melbourne_model.predict(X.head()))
print("The actual values are :")
print(y.head())

#
# and save this all to the dataset
# along with the original values
#
Xout = X.copy()
Xout.insert(5, "Price", y)
Xout.insert(6, "PredictedPrice", melbourne_model.predict(X))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6.5, 6.5))

sns.despine(fig, left=True, bottom=True)
sns.scatterplot(x="Price", y="PredictedPrice",
                hue="Landsize", size="Rooms",
                palette="ch:r=-.2,d=.3_r",
                sizes=(5,40), linewidth=0,
                data=Xout, ax=ax)
ax.semilogx()
ax.semilogy()
```
